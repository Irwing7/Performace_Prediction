```python
import pandas as pd
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df = df.drop(columns = ['Employee_ID','Hire_Date','Sick_Days', 'Remote_Work_Frequency', 'Team_Size','Work_Hours_Per_Week','Training_Hours','Resigned'])
```

```python
y = df["Performance_Score"]
x = df[["Years_At_Company","Monthly_Salary","Overtime_Hours","Promotions","Employee_Satisfaction_Score"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
import joblib as jb
jb.dump(scaler,"scaler.pkl")
```

```python
from sklearn.metrics import accuracy_score
def model_performance(predictions):
  print("Accuracy score is {}".format(accuracy_score(y_test,predictions)))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(x_train_scaled,y_train)
```

```python
model_predictions = model.predict(x_test_scaled)
model_performance(model_predictions)
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors":[3,5,6,9,13],"weights": ["distance"]}
```

```python
grid_knn = GridSearchCV(KNeighborsClassifier(),param_grid,cv = 5)
```

```python
grid_knn.fit(x_train_scaled,y_train)
```

```python
grid_knn.best_params_
```

```python
predictions = grid_knn.predict(x_test_scaled)
model_performance(predictions)
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
param_grid = {"C":[0.01,0.1,0.5,1],"kernel":["linear","rbf","poly"]}
```

```python
grid_svc = GridSearchCV(SVC(),param_grid,cv = 2)
```

```python
grid_svc.fit(x_train_scaled,y_train)
```

```python
predicts = grid_svc.predict(x_test_scaled)
model_performance(predicts)
```
